```python
import doctest

def recherche_dicho(T, e):
    """
    >>> recherche_dicho([0, 1, 2, 3, 4], 3)
    True
    >>> recherche_dicho([0, 1, 3, 6], 2.5)
    False
    """
    
    if len(T) == 0: return False
    if len(T) == 1: return e == T[0]
    indice_milieu = len(T) // 2
    if T[indice_milieu] == e:
        return True
    elif T[indice_milieu] > e:
        return recherche_dicho(T[:indice_milieu], e)
    else:
        return recherche_dicho(T[indice_milieu+1:], e)
    

def recherche(T, e):
    """
    >>> recherche([1, 8, 3.5], 3.5)
    True
    >>> recherche([0, 4, 7.8, 5], 2)
    False
    """
    if len(T) == 0: return False
    elif len(T) == 1: return T[0] == e

    return T[0] == e or recherche(T[1:], e)
    
def recherche2(T, e):
    """
    >>> recherche2([1, 8, 3.5], 3.5)
    True
    >>> recherche2([0, 4, 7.8, 5], 2)
    False
    """
    if len(T) == 0:
        return False
    elif len(T) == 1:
        return e == T[0]
    
    indice_milieu = len(T) // 2
    return recherche2(T[:indice_milieu], e) or \
           recherche2(T[indice_milieu:], e)
        
doctest.testmod()
    
```
