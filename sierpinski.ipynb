```python
__author__ = "david cobac"

from PIL import Image, ImageDraw


def milieu(A, B):
    return tuple([(a + b) / 2 for a, b in zip(A, B)])
    


def sierpinski(contexte, triangle, niveau):
    if niveau == 1:
        contexte.polygon(triangle)
    else:
        A, B, C = triangle
        D = milieu(A, B)
        E = milieu(A, C)
        F = milieu(B, C)
        sierpinski(contexte, [A, D, E], niveau - 1)
        sierpinski(contexte, [D, B, F], niveau - 1)
        sierpinski(contexte, [E, F, C], niveau - 1)


W = 1_000
H = W
im = Image.new("RGB", (W, H), "black")
ctx_g = ImageDraw.Draw(im)

ABC = [(0, H - 1), (W - 1, H - 1), (W / 2, 0)]
N = 10

sierpinski(ctx_g, ABC, N)

im.save("sierpinski.png")
```
